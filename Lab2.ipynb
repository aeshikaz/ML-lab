```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_breast_cancer

data = load_breast_cancer()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target

# Artificially introduce missing values for demonstration
df.iloc[0:5, 0] = np.nan
df.iloc[2:4, 1] = np.nan

print("Missing values before fillna:")
print(df.isnull().sum())

# Fill missing numerical values with mean
df['mean radius'] = df['mean radius'].fillna(df['mean radius'].mean())

# Fill missing numerical values with median
df['mean texture'] = df['mean texture'].fillna(df['mean texture'].median())

print("\nMissing values after fillna:")
print(df.isnull().sum())
```

```python
import pandas as pd
from sklearn.datasets import load_breast_cancer

data = load_breast_cancer()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target

print("Before renaming:")
print(df.columns.tolist())

df = df.rename(columns={
    'mean radius': 'Radius',
    'mean texture': 'Texture',
    'mean perimeter': 'Perimeter',
    'mean area': 'Area',
    'mean smoothness': 'Smoothness'
})

print("\nAfter renaming:")
print(df.columns.tolist())
print(df.head())
```

```python
import pandas as pd
from sklearn.datasets import load_breast_cancer

data = load_breast_cancer()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target

# Introduce inconsistent values for demonstration
df.loc[0, 'mean radius'] = -99
df.loc[1, 'mean radius'] = -99

print("Before replacing incorrect values:")
print(df['mean radius'].head())

# Replace incorrect value (-99) with column mean
mean_val = df[df['mean radius'] != -99]['mean radius'].mean()
df['mean radius'] = df['mean radius'].replace(-99, mean_val)

print("\nAfter replacing incorrect values:")
print(df['mean radius'].head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer

data = load_breast_cancer()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target

# Line plot - mean radius across first 50 samples
plt.figure(figsize=(10, 4))
plt.plot(df['mean radius'][:50])
plt.title("Line Plot - Mean Radius (First 50 Samples)")
plt.xlabel("Sample Index")
plt.ylabel("Mean Radius")
plt.show()

# Bar plot - count of each target class
plt.figure(figsize=(6, 4))
df['target'].value_counts().plot(kind='bar', color=['salmon', 'steelblue'])
plt.title("Bar Plot - Target Class Distribution")
plt.xlabel("Class (0=Malignant, 1=Benign)")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer

data = load_breast_cancer()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target

plt.figure(figsize=(7, 5))
plt.scatter(df['mean radius'], df['mean area'], c=df['target'], cmap='coolwarm', alpha=0.6)
plt.title("Scatter Plot - Mean Radius vs Mean Area")
plt.xlabel("Mean Radius")
plt.ylabel("Mean Area")
plt.colorbar(label='Target (0=Malignant, 1=Benign)')
plt.show()

print("Interpretation: There is a strong positive relationship between")
print("mean radius and mean area — as radius increases, area increases.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer

data = load_breast_cancer()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target

plt.figure(figsize=(7, 5))
plt.hist(df['mean radius'], bins=20, color='steelblue', edgecolor='black')
plt.title("Histogram - Mean Radius Distribution")
plt.xlabel("Mean Radius")
plt.ylabel("Frequency")
plt.show()

print("Skewness of mean radius:", df['mean radius'].skew())
print("Interpretation: Positive skew means most values are on the lower side")
print("with a few large outliers pulling the tail to the right.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer

data = load_breast_cancer()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target

plt.figure(figsize=(12, 5))
df[['mean radius', 'mean texture', 'mean perimeter', 'mean area', 'mean smoothness']].boxplot()
plt.title("Boxplot - Numerical Features")
plt.ylabel("Values")
plt.xticks(rotation=15)
plt.show()

print("Interpretation: Points beyond the whiskers are potential outliers.")
print("mean area and mean perimeter show the most spread and outliers.")
```
